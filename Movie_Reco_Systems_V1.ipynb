```python
movies = df_movies.merge(df_credits, on = 'title')
movies.head(1)
```

```python
# Genres, homepage, id, keywords, title, overview, cast and crew

movies = movies[['movie_id','title', 'overview', 'genres', 'keywords','cast', 'crew']]
```

```python
movies.head()
```

```python
# Removing the missing value
movies.isnull().sum()
```

```python
movies.dropna(inplace = True)
```

```python
movies.isnull().sum()
```

```python
movies.iloc[0].genres
```

```python
# preprocessing
#[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]
#['Action', 'Adventure', 'Fantasy' and more]
```

```python
import ast
ast.literal_eval
```

```python
def convert(obj):
  l = []
  for i in ast.literal_eval(obj):
      l.append(i['name'])
  return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert2(obj):
  l = []
  counter = 0
  for i in ast.literal_eval(obj):
      if counter != 3:
          l.append(i['name'])
          counter += 1
      else:
          break
  return l
```

```python
movies['cast'] = movies['cast'].apply(convert2)
```

```python
def fetch_director(obj):
  l = []
  for i in ast.literal_eval(obj):
      if i['job'] == 'Director':
          l.append(i['name'])
          break
  return l
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head()
```

```python
# removing spaces from the list
#Sam Worthington to SamWorthington to make the tags in shape
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies['tags'].iloc[0]
```

```python
new_df = movies[['movie_id', 'title', 'tags']]
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
# doing steming for this data ['loving', 'lover', 'loved']
# result we will get after steaming['love', 'love', 'love']
```

```python
import nltk
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
ps.stem('loved'), ps.stem('doing'), ps.stem('killing')
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return  " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
# its showing the similarity of the 1st movie
similarity[0]
```

```python
similarity.shape
```

```python
new_df[new_df['title'] == 'Batman Begins'].index[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse = True, key=lambda x:x[1])[1:11]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse = True, key=lambda x:x[1])[1:11]

    for i in movies_list:
        print(i[0])
```

```python
recommend('Avatar')
```
